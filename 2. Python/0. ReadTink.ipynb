```python
def getLightsData(data_loaded):

    play = pd.DataFrame(data_loaded)

    for i in range(len(play.categories)-1):

        
        frame =play.categories[i]

        kategori = list(play.categories[i].keys())[i]
        #print(kategori)

        # Rename keys by adding 'Gambling' prefix and exclude 'category'
        data_renamed = {f'{kategori}_{k}': v for k, v in frame.items() if k != 'category'}

        if i == 0:
            df =  pd.DataFrame([data_renamed])
        else:
            df = pd.concat([df,pd.DataFrame([data_renamed])],axis=1)

    return df



def getCategoryAnalysis(fr):
    
    df = pd.DataFrame(fr['categoryAnalysis'])
    # Use pivot_table and specify an aggregation function, 'first' will pick the first non-null value per category
    pivoted_df = df.pivot_table(index=[df.index % 1], columns='category', values=['avg', 'avg3Months', 'avg6Months', 'count', 'sum', 'trend'], aggfunc='first')

    # Flatten the columns to create a single row DataFrame
    pivoted_df.columns = ['_'.join(col).strip() for col in pivoted_df.columns.values]

    pivoted_df.reset_index(drop=True, inplace=True)

    # Reset index to get everything in one row
    return pivoted_df



def getCreditorAnalysis(fr):
    data = fr['creditorAnalysis']

    # Convert the nested dictionary to a DataFrame
    rows = []
    for category, entries in data.items():
        for entry in entries:
            entry['category'] = category
            rows.append(entry)
    df = pd.DataFrame(rows)

    # Group by category and calculate total sum and count
    category_summary = df.groupby('category').agg({'sum': 'sum', 'count': 'sum'})

    # Pivot the table to make categories into columns and combine metrics into column names
    final_df = category_summary.unstack().to_frame().T
    final_df.columns = [' '.join(col).strip() for col in final_df.columns.values]

    return final_df


def getIncomeAnalysis(fr):

    # Convert to DataFrame
    df = pd.DataFrame(fr['incomeAnalysis'])

    # Convert date strings to datetime objects
    df['date'] = pd.to_datetime(df['date'])

    # Sort DataFrame by date in descending order
    df = df.sort_values('date', ascending=False)

    # Create a column for the month index where 0 is the most recent
    df['month_index'] = range(len(df))

    return df



def getSpendAnalysis(fr):

    result_df = pd.DataFrame()  

    for c in range(len(fr['spendAnalysis'])):
        df = pd.DataFrame([fr['spendAnalysis'][c]])
        
        current_category = df.iloc[0, 0] 

        # Append the current category to each column name
        df.columns = [f"{current_category}_{col}" for col in df.columns]

        # Concatenate horizontally
        if result_df.empty:
            result_df = df
        else:
            result_df = pd.concat([result_df, df], axis=1)
            
    return result_df

```

```python
import numpy as np
import pandas as pd
import pyodbc  # using pyodbc as the database connector
import json

class DataPreprocessor:
    
    def __init__(self, server, database, username, password, driver):
        self.server = server
        self.database = database
        self.username = username
        self.password = password
        self.driver = driver  # Add a driver attribute, since pyodbc requires specifying the ODBC driver to use

    def _connect_to_db(self):
        # Construct the connection string
        conn_str = f'DRIVER={self.driver};SERVER={self.server};DATABASE={self.database};UID={self.username};PWD={self.password}'
        # Connect to the database using pyodbc
        self.conn = pyodbc.connect(conn_str)

    def fetch_data_from_sql(self, sql_query, params):
        try:
            # Ensure the database connection is established
            self._connect_to_db()
            
            # Check if sql_query is not empty or None
            if not sql_query:
                raise ValueError("SQL query is empty or None.")
            
            # Use pd.read_sql to execute the query and load the results into a DataFrame
            df = pd.read_sql(sql_query, con=self.conn, params=params)
            
            # Check if DataFrame is empty
            if df.empty:
                raise ValueError("Query returned no results.")
            
            return df
        except Exception as e:
            print(f"Error fetching data: {e}")
            # Optionally, re-raise the exception if you want calling code to handle it
            raise
        finally:
            # Close the database connection
            self.conn.close()

# Define the SQL query with parameter placeholder for SSN

sql_query = """
WITH Account AS (
    SELECT [AccountJsonID],
           [TransactionJsonID],
           [ReportDate],
           [SSN],
           [AccountJSON],
           [TransactionJSON],
           [CheckID],
           [QueryString]
    FROM [nim-production-backup].[tink].[TinkReports]
    WHERE SSN = ?
),
LastReport AS (
    SELECT MAX(ReportDate) AS ReportDate
    FROM Account
)
SELECT * 
FROM Account AS a
INNER JOIN LastReport AS l ON a.ReportDate = l.ReportDate
"""
```
